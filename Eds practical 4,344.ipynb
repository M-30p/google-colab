```python
sal=list()
title=list()
sal_usd=list()
emp_type=list()
experience=list()
per=list()
f=open("company1.csv","r")
data=f.readline()
while(True):

  data=f.readline()
  if not data:
    break;
  print(data)
  temp=data.split(",")
  sal.append(int(temp[4]))
  title.append(temp[3])
  sal_usd.append(int(temp[5]))
  emp_type.append(temp[2])
  experience.append(temp[1])
```

1.What is the highest paid job title?

```python
print(sal,"\n\n")

print(title[sal.index(max(sal))],"is Job title of highest paid employe")
```

2.What is the highest salary in USD?

```python
print(sal_usd,"\n\n")

print(max(sal_usd),"USD is highest salary\n\n")
```

3.What is the count of part time employees?

```python
print(emp_type,"\n\n")
frequency ={}
for emp in emp_type:
  if emp in frequency:
    frequency[emp]+=1
  else:
    frequency[emp]=1
print(frequency,"\n\n")

print(frequency['Part'],"employees are part time employeees")
```

4.What is the count of full time employees?

```python
print(frequency['Full'],"employees are full time employeees\n\n")
```

5.What is the count of newly joined emplpyees?

```python
print(experience,"\n\n")
frequency ={}
for level in experience:
  if level in frequency:
    frequency[level]+=1
  else:
    frequency[level]=1
print(frequency,"\n\n")

print(frequency['New'],"employees are newly joined employees")


```
