```python
import numpy as np
import pandas as pd

data=pd.read_csv('/content/coffee.csv')
```

What is the average bag weight of coffee beans across all countries?



```python
average_bag_weight = data['Bag Weight'].mean()

print("The average bag weight of coffee beans is ",round(average_bag_weight,2),"kg.")
```

17: What is the total aftertaste rating of coffee beans from Thailand with more than 4 category two defects?



```python
thailand_high_defects_data = data[(data['Country of Origin'] == 'Thailand') & (data['Category Two Defects'] > 4)]
total_aftertaste_thailand_high_defects = thailand_high_defects_data['Aftertaste'].sum()

print("The total aftertaste rating of coffee beans from Thailand with more than 4 category two defects is",total_aftertaste_thailand_high_defects)
```

2: Which country has the highest number of bags of coffee beans?

```python
country_with_highest_bags = data.groupby('Country of Origin')['Number of Bags'].sum().idxmax()

print("The country with the highest number of bags of coffee beans is",country_with_highest_bags)
```

**Machine** **learning** **line** **of** **regression**.

---







```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/content/result2.csv')
print(data.shape)
data.head()
```

```python
X=data['Maths'].values
Y=data['English'].values

mean_X=np.mean(X)
mean_Y=np.mean(Y)

n=len(X)

num=0
den=0
for i in range (n):
    num+=(X[i]-mean_X) * (Y[i]-mean_Y)  #summation of (x-x')*(y-y')
    den+=(X[i]-mean_X) ** 2             #summation of (x-x')^2
m=num/den
c=mean_Y-(m* mean_X)

print(m)
print(c)
```

```python
max_x=np.max(X)+100
min_x=np.min(X)-100

x=np.linspace(min_x,max_x,1000)
y=m*x+c

plt.plot(x,y,color='g',label='Regression line')
plt.scatter(X,Y,color='b',label='Scatter plot')

plt.xlabel('Maths')
plt.ylabel('English')
plt.legend()
plt.show()
```

```python
sum1=0
sum2=0

for i in range(n):
    yp=m*X[i]+ c
    sum1+=(Y[i]-yp) **2
    sum2+=(Y[i]-mean_Y)**2
r2=1-(sum1/sum2)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X=X.reshape((n,1))

reg=LinearRegression()

reg=reg.fit(X,Y)

r2_score=reg.score(X,Y)
print(r2_score)
```

**KNN Classification or clustering**

---



```python
import matplotlib.pyplot as plt
x =[8.42,8.25,8.17,8.08,7.92,8.17,8.25,8.08,8,8]
y =[35,80,25,22,24,30,27,90,30,60]

plt.scatter(x,y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x,y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias , marker ='o')
plt.title('Elbow method')
plt.xlabel('Numbers of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
